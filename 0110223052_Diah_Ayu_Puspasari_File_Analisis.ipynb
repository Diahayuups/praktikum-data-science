# DATA PREPARATION

## Merge seluruh Data

Pada tahap ini, data order_df digabung dengan users_df menggunakan kolom user_id sebagai kunci penghubung. Metode left join digunakan agar semua data transaksi tetap dipertahankan meskipun tidak semua memiliki data pengguna yang lengkap. Hasil penggabungan ini disimpan dalam variabel orders_users.

Kedua, hasil penggabungan tersebut kemudian digabungkan lagi dengan product_df menggunakan kolom product_id. Proses ini menghasilkan satu dataset utuh yang dinamakan mydata, yang sudah memuat informasi lengkap dari transaksi, pengguna, dan produk. Dataset mydata ini akan menjadi pusat analisis utama dalam menjawab seluruh pertanyaan UAS.

```python
# Gabung order dan user
orders_users = order_df.merge(
    users_df,
    how='left',
    left_on='user_id',
    right_on='id',
    suffixes=('', '_user')
)

# Gabung dengan product
mydata = orders_users.merge(
    product_df,
    how='left',
    left_on='product_id',
    right_on='id',
    suffixes=('', '_product')
)
```

## Cek Missing Value

```python
# 1. Cek missing values
missing = mydata.isnull().sum().sort_values(ascending=False)
missing
```

```python
# 1. Isi city kosong dengan 'Unknown'
mydata['city'] = mydata['city'].fillna('Unknown')

# 2. Isi brand kosong dengan 'No Brand'
mydata['brand'] = mydata['brand'].fillna('No Brand')

# 3. Isi name kosong dengan 'Unnamed Product'
mydata['name'] = mydata['name'].fillna('Unnamed Product')

# 4. Untuk kolom datetime (returned_at, delivered_at, shipped_at), kita biarkan saja
# karena NaT (missing datetime) artinya belum terjadi, misalnya belum dikirim atau dikembalikan
```

## Menghapus Kolom

```python
mydata = mydata.drop(columns=[
    'id', 'id_user', 'id_product', 'inventory_item_id',
    'email', 'sku', 'distribution_center_id', 'created_at_user'
])
```

Dalam tahap pembersihan data, beberapa kolom dihapus dari dataset gabungan (mydata) untuk menyederhanakan struktur dan menghilangkan redundansi informasi. Kolom id_user dan id_product dihapus karena merupakan duplikasi dari user_id dan product_id yang sudah cukup mewakili identitas pengguna dan produk. Kolom id juga dihapus karena berasal dari proses penggabungan dan tidak memiliki arti khusus dalam analisis. Selain itu, created_at_user yang menunjukkan waktu pendaftaran pengguna dianggap tidak relevan karena fokus analisis diarahkan pada waktu transaksi, bukan waktu registrasi.

Kolom email dihapus karena tidak diperlukan untuk menjawab pertanyaan analisis dan tidak memberikan insight tambahan. Sementara itu, kolom seperti sku (kode stok produk) dan distribution_center_id (kode gudang distribusi) juga tidak digunakan dalam proses analisis ini karena lebih berkaitan dengan aspek logistik dan manajemen inventaris. Dengan menghapus kolom-kolom tersebut, dataset menjadi lebih ringkas, efisien, dan hanya memuat informasi yang benar-benar relevan untuk analisis lebih lanjut.

# Jawab Pertanyaan

## No 1. Dari data pengguna, ada berapa banyak yang berasal dari Korea Selatan?


```python
# Filter user dari Korea Selatan
korea_users = mydata[mydata['country'] == 'South Korea']

# Ambil user unik berdasarkan user_id
jumlah_user_korsel = korea_users['user_id'].nunique()

print("Jumlah pengguna dari Korea Selatan:", jumlah_user_korsel)
```

## No 2. Berapa banyak pelanggan pria dan wanita dari Brandenburg, Jerman?

```python
# Filter pengguna yang berasal dari Brandenburg, Jerman
brandenburg = mydata[
    (mydata['state'] == 'Brandenburg') &
    (mydata['country'] == 'Germany')
]

# Ambil hanya user unik
brandenburg_unique = brandenburg.drop_duplicates(subset='user_id')

# Hitung berdasarkan gender
jumlah_gender = brandenburg_unique['gender'].value_counts()

print(jumlah_gender)
```

## No. 3 Berapa usia pelanggan termuda dan tertua? Dari negara mana kah mereka?

```python
# Ambil hanya user unik
unique_users = mydata.drop_duplicates(subset='user_id')

# Usia termuda
termuda = unique_users.loc[unique_users['age'].idxmin(), ['age', 'country']]

# Usia tertua
tertua = unique_users.loc[unique_users['age'].idxmax(), ['age', 'country']]

print("Usia Termuda:", termuda['age'], "dari", termuda['country'])
print("Usia Tertua:", tertua['age'], "dari", tertua['country'])
```

## No. 4 Sebutkan nama 5 pengguna yang terdaftar paling baru.

```python
# Ambil user unik dulu
unique_users = mydata.drop_duplicates(subset='user_id')

unique_users = unique_users.copy()
unique_users['created_at'] = pd.to_datetime(unique_users['created_at'], errors='coerce')

# Urutkan berdasarkan tanggal daftar terbaru
pengguna_baru = unique_users.sort_values(by='created_at', ascending=False)

# Ambil 5 nama paling baru
top5 = pengguna_baru[['first_name', 'last_name', 'created_at']].head(5)

print(top5)
```

## No. 5 Ada berapa macam kategori produk yang dijual oleh Looker?

```python
# Hitung jumlah kategori unik
jumlah_kategori = mydata['category'].nunique()

# Tampilkan daftar kategori
daftar_kategori = mydata['category'].unique()

print("Jumlah kategori produk:", jumlah_kategori)
print("Kategori-kategori produk:")
print(daftar_kategori)
```

## No. 6 Produk apa yang paling banyak terjual di 2020? Ada berapa transaksi penjualan produk tersebut?

```python
# Pastikan kolom tanggal sudah dalam format datetime
mydata['created_at'] = pd.to_datetime(mydata['created_at'], errors='coerce')

# Filter transaksi tahun 2020
data_2020 = mydata[mydata['created_at'].dt.year == 2020]

# Hitung jumlah transaksi per produk
produk_terjual = data_2020['product_id'].value_counts()

# Ambil produk dengan jumlah transaksi terbanyak
produk_top_id = produk_terjual.idxmax()
jumlah_transaksi = produk_terjual.max()

# Ambil nama produknya
nama_produk = mydata[mydata['product_id'] == produk_top_id]['name'].iloc[0]

print("Produk paling banyak terjual di tahun 2020:")
print("Nama Produk:", nama_produk)
print("Jumlah Transaksi:", jumlah_transaksi)
```

## No. 7 Berapa banyak transaksi yang dibatalkan dari tahun 2019 ke tahun 2022? Bagaimana pendapatmu tentang informasi tersebut?

```python
# Pastikan kolom returned_at sudah datetime
mydata['returned_at'] = pd.to_datetime(mydata['returned_at'], errors='coerce')

# Ambil hanya transaksi yang dikembalikan
retur = mydata[mydata['returned_at'].notnull()]

# Ambil tahun dari kolom returned_at
retur['tahun_retur'] = retur['returned_at'].dt.year

# Hitung jumlah retur per tahun 2019-2022
retur_per_tahun = retur[retur['tahun_retur'].between(2019, 2022)]
hasil = retur_per_tahun['tahun_retur'].value_counts().sort_index()

print("Jumlah transaksi dibatalkan per tahun (2019-2022):")
print(hasil)
```

## No. 8 Buatkan tabel distribusi pelanggan berdasarkan sumber trafiknya. Yang ingin diketahui adalah channel pemasaran yang mana yang menghasilkan pelanggan paling banyak.

```python
# Ambil user unik
user_unik = mydata.drop_duplicates(subset='user_id')

# Hitung jumlah user per traffic_source
distribusi_channel = user_unik['traffic_source'].value_counts()

print("Distribusi pelanggan berdasarkan channel pemasaran:")
print(distribusi_channel)
```

```python
# Hitung distribusi per negara dan traffic_source
distribusi_per_negara = user_unik.groupby(['country', 'traffic_source'])['user_id'].count().unstack().fillna(0)

print("Distribusi pelanggan per negara berdasarkan channel:")
print(distribusi_per_negara)
```

# Insight Tambahan

## Channel Pemasaran Terfavorit di Setiap Negara

Untuk mengetahui channel pemasaran yang paling efektif di setiap negara.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Ambil user unik
user_unik = mydata.drop_duplicates(subset='user_id')

# Hitung distribusi channel per negara
channel_per_negara = user_unik.groupby(['country', 'traffic_source'])['user_id'].count().unstack().fillna(0)

# Ambil top 10 negara berdasarkan total pengguna
top_negara = channel_per_negara.sum(axis=1).sort_values(ascending=False).head(10).index

# Plot hanya top 10 negara
channel_per_negara.loc[top_negara].plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Distribusi Channel Pemasaran di 10 Negara Teratas')
plt.ylabel('Jumlah Pengguna')
plt.xlabel('Negara')
plt.legend(title='Channel')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Negara dengan Jumlah Pengguna Terbanyak

Untuk mengetahui negara dengan basis pengguna terbesar.

```python
# Hitung jumlah user unik per negara
user_per_country = user_unik['country'].value_counts().head(10)

# Plot
user_per_country.plot(kind='bar', figsize=(10,5), color='skyblue')
plt.title('10 Negara dengan Jumlah Pengguna Terbanyak')
plt.ylabel('Jumlah Pengguna')
plt.xlabel('Negara')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Perbandingan Jumlah Return Tiap Tahun

Untuk melihat apakah tren pengembalian produk meningkat atau menurun.

```python
# Pastikan kolom returned_at datetime
mydata['returned_at'] = pd.to_datetime(mydata['returned_at'], errors='coerce')

# Ambil tahun return
retur_tahunan = mydata[mydata['returned_at'].notnull()]
retur_tahunan['tahun'] = retur_tahunan['returned_at'].dt.year

# Hitung jumlah retur per tahun
retur_per_tahun = retur_tahunan['tahun'].value_counts().sort_index()

# Plot
retur_per_tahun.plot(kind='bar', color='salmon')
plt.title('Jumlah Retur Produk per Tahun')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Transaksi Retur')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
