```python
# membaca file csv penilaian 2019
data2019 = pd.read_csv(path + '/karyawan-penilaian2019.csv')
data2019
```

```python
# membaca file csv penilaian 2018
data2018 = pd.read_csv(path + '/karyawan-penilaian2018.csv')
data2018
```

```python
# Rename kolom dulu biar seragam
data2018 = data2018.rename(columns={
    'ID': 'id_karyawan',
    'Nama Pegawai': 'nama_pegawai'
})
```

```python
# Gabungkan data dengan outer join berdasarkan id_karyawan dan nama_pegawai
data_gabungan = pd.merge(data2018, data2019, on=['id_karyawan', 'nama_pegawai'], how='outer')
data_gabungan
```

## Siapa saja karyawan yang performanya naik dari 2018 ke 2019?

```python
# Filter hanya yang performanya naik
performa_naik = data_gabungan[data_gabungan['Penilaian 2019'] > data_gabungan['Penilaian 2018']]

performa_naik
```

```python
# membuat kolom baru untuk Keterangan karyawan yang baru dan keluar
def status_karyawan(row):
    if pd.isnull(row['Penilaian 2018']) and not pd.isnull(row['Penilaian 2019']):
        return 'Karyawan Baru'
    elif not pd.isnull(row['Penilaian 2018']) and pd.isnull(row['Penilaian 2019']):
        return 'Karyawan Keluar'
    elif not pd.isnull(row['Penilaian 2018']) and not pd.isnull(row['Penilaian 2019']):
        return 'Tetap'
    else:
        return 'Tidak Ada Data'

data_gabungan['Keterangan'] = data_gabungan.apply(status_karyawan, axis=1)
data_gabungan
```

## Siapa saja karyawan baru di 2019?

```python
# menampilkan data karyawan baru
karyawan_baru = data_gabungan[data_gabungan['Keterangan'] == 'Karyawan Baru']
karyawan_baru
```

## Siapa saja yang keluar di 2019?

```python
# menampilkan data karyawan yang keluar
karyawan_keluar = data_gabungan[data_gabungan['Keterangan'] == 'Karyawan Keluar']
karyawan_keluar
```

## Table gabungannya di simpan sebagai file CSV
