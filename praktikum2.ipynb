Penjelasan Kolom:
*   Unnamed: 0 → Indeks, tidak memiliki makna analitis dan bisa dihapus.
*   CRIM → Tingkat kejahatan per kapita di kota (semakin tinggi, semakin rendah harga rumah)
*   ZN → Persentase luas tanah untuk perumahan besar (>25.000 sqft).
*   INDUS → Persentase area yang digunakan untuk bisnis non-ritel.
*   CHAS → Dummy variable (1 = berbatasan dengan sungai Charles, 0 = tidak).
*   NOX → Konsentrasi polusi udara (nitrogen oksida) dalam ppm.
*   RM → Rata-rata jumlah kamar per rumah (semakin tinggi, semakin mahal rumah).
*   AGE → Persentase rumah yang dibangun sebelum tahun 1940.
*   DIS → Jarak rata-rata ke pusat pekerjaan utama di Boston.
*   RAD → Indeks aksesibilitas ke jalan raya utama.
*   TAX → Tarif pajak properti per $10.000.
*   PTRATIO → Rasio jumlah murid terhadap guru di kota (semakin tinggi, semakin kurang optimal pendidikan).
*   B → Proporsi penduduk keturunan Afrika-Amerika (dihitung dengan rumus tertentu).
*   LSTAT → Persentase penduduk dengan status ekonomi rendah.
*   MEDV → Median nilai properti dalam ribuan dolar (sering dijadikan target prediksi).
